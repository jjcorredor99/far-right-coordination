```python
import tweepy
import pandas as pd
```

```python
APP_KEY3 = "xxxxxxxxxxxxxxxxxxxx"
APP_SECRET3 =  "xxxxxxxxxxxxxxxxxxxx"
OAUTH_TOKEN3 =  "xxxxxxxxxxxxxxxxxxxx"
OAUTH_TOKEN_SECRET3 = "xxxxxxxxxxxxxxxxxxxx"
```

```python
auth = tweepy.OAuthHandler(APP_KEY3, APP_SECRET3)
auth.set_access_token(OAUTH_TOKEN3, OAUTH_TOKEN_SECRET3)
api3 = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python

datos_v3= pd.read_excel("./bases/uribistas_V3.xlsx")
datos_v4= pd.read_excel("./bases/uribistas_V4.xlsx")
```

```python
datos_v4=datos_v4.sort_values("grupo")
datos_v4=datos_v4.dropna(subset=["grupo","cuenta"])

datos_v3=datos_v3.sort_values("grupo")
datos_v3=datos_v3.dropna(subset=["grupo","cuenta"])
print(len(datos_v3),len(datos_v4))
```

```python
cuenta_v3=set(datos_v3["cuenta"])
cuenta_v4=set(datos_v4["cuenta"])

accounts=cuenta_v4 - cuenta_v3
```

```python
datos=datos.sort_values("grupo")
datos2=datos.dropna(subset=["grupo","cuenta"])
datos2
```

```python
cuentas=datos2["cuenta"]
len(cuentas)
```

```python
seguidores=[]
exceptions=[]
for i in accounts:
    
    try:
        user_info=api3.get_user(i)._json
        seguidores.append(user_info)
    except Exception as e:
        exceptions.append( {i: e})
```

```python
len(exceptions)
```

```python
informacion=pd.DataFrame(seguidores)
informacion
```

```python
info_vieja= pd.read_excel("./bases/informacion_politicos.xlsx")
```

```python
info= pd.concat([info_vieja, informacion], ignore_index=True )
info
```

```python
info.to_excel("./bases/informacion_politicos_V2.xlsx")
```
